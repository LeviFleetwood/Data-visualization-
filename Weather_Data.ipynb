```python
df = tempDataFull.groupby(['Day','Month'],as_index=False)['AvgTemperature'].min()
df
```

```python
valid_size_values = df['AvgTemperature'].loc[df['AvgTemperature'] >= 0]
valid_df = df.loc[df['AvgTemperature'] >= 0]
fig = px.scatter(valid_df, x='Month', y='AvgTemperature',
                 animation_frame='Day', size=valid_size_values,
                 range_x=[-1, 13], range_y=[30, 100], color='AvgTemperature',
                 title='Min temp for each day of each month from 1995-2014')
fig['layout'].pop('updatemenus')
fig.show()
```
